```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/list/ls080044434/")
page
```

```python
con=BeautifulSoup(page.content)
con
```

```python
title=con.find('h3',class_="lister-item-header")
title
```

```python
title.text
```

```python
title.text.split()[2:6]
```

```python
rating=con.find('div',class_="ipl-rating-star small")
rating
```

```python
rating.text.split()[:10]
```

```python
yor=con.find('span',class_="lister-item-year text-muted unbold")
yor
```

```python
yor.text
```

```python
titles=[]

for i in con.find_all('h3',class_="lister-item-header"):
    titles.append(i.text.split()[2:6])
titles
```

```python
rating=[]
for i in con.find_all('div',class_="ipl-rating-star small"):
    rating.append(i.text.split()[:10])
```

```python
rating
```

```python
yor=[]
for i in con.find_all('span',class_="lister-item-year text-muted unbold"):
    yor.append(i.text)
yor
```

```python
df=pd.DataFrame({'Title':titles,'Year Of Release':yor,'Rating':rating})
df
```
